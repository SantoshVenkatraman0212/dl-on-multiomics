```python
# Load the datasets
gleason_df = pd.read_csv('Balanced_Gleason_Score_Dataset.csv')
genefo = pd.read_csv('Mutated_Genes.csv')

# Extract gene names by removing the "omics" prefixes (GE_, CNA_, DM_)
# List of unique gene names
gene_columns = [col for col in gleason_df.columns if col.startswith(('GE_', 'CNA_', 'DM_'))]
genes = list(set([col.split('_')[1] for col in gene_columns]))

gleason_long = []

for gene in genes:
    # Create a dictionary to hold gene-specific data
    gene_data = {
        'Gene': gene,
    }
    
    # Check and extract GE, CNA, DM values if they exist
    if f'GE_{gene}' in gleason_df.columns:
        gene_data['GE'] = gleason_df[f'GE_{gene}']
    else:
        gene_data['GE'] = pd.Series([None] * len(gleason_df))  # Fill with None if not present

    if f'CNA_{gene}' in gleason_df.columns:
        gene_data['CNA'] = gleason_df[f'CNA_{gene}']
    else:
        gene_data['CNA'] = pd.Series([None] * len(gleason_df))  # Fill with None if not present

    if f'DM_{gene}' in gleason_df.columns:
        gene_data['DM'] = gleason_df[f'DM_{gene}']
    else:
        gene_data['DM'] = pd.Series([None] * len(gleason_df))  # Fill with None if not present

    # Convert the dictionary to a DataFrame and append to the list
    gene_df = pd.DataFrame(gene_data)
    gleason_long.append(gene_df)

# Combine all gene DataFrames into a single DataFrame
gleason_long = pd.concat(gleason_long, ignore_index=True)

# Merge with the second dataset based on the gene names
merged_df = pd.merge(gleason_long, genefo, on='Gene', how='inner')

print(merged_df.head)
merged_df.to_csv('Combined_Omics.csv', index=False)
```

```python
# Tried Autoencoder stuff on revised dataset.

from tensorflow import keras
from tensorflow.keras import layers

# Prepare the input data

combined_data = pd.read_csv('Combined_Omics.csv')
X = combined_data.drop(columns=['Gene', 'MutSig(Q-value)', '# Mut', '#Profiled Samples', 'Freq', 'Is Cancer Gene (source: OncoKB)']).values

# Handle missing values: Replace NaN with 0
X = np.nan_to_num(X)

# Define the autoencoder architecture
input_dim = X.shape[1]  # Number of features (GE, CNA, DM)
encoding_dim = 32  # Latent space dimensionality

# Build the model
input_layer = layers.Input(shape=(input_dim,))
encoder = layers.Dense(encoding_dim, activation='relu')(input_layer)
decoder = layers.Dense(input_dim, activation='sigmoid')(encoder)

autoencoder = keras.Model(input_layer, decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Train the autoencoder
autoencoder.fit(X, X, epochs=50, batch_size=32, shuffle=True, validation_split=0.2)

# Extract the encoder to get the latent space representation
encoder_model = keras.Model(inputs=input_layer, outputs=encoder)

# Get the latent space representation for the data
latent_space = encoder_model.predict(X)

# Convert latent space to a DataFrame for further analysis
latent_df = pd.DataFrame(latent_space, columns=[f'Latent_{i}' for i in range(encoding_dim)])

# Display the latent space DataFrame
print(latent_df.head())
```
